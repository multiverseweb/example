```python
import numpy as np
m=int(input("Length of land: "))
n=int(input("Width of land: "))
random_array = np.random.randint(0, 10, (m, n))
land = random_array.tolist()

peaks=[]
valley=[]
for i in range(1,m-1):
    for j in range(1,n-1):
        point=land[i][j]
        surroundings=[land[i][j-1],land[i][j+1],land[i-1][j-1],land[i-1][j],land[i-1][j+1],land[i+1][j-1],land[i+1][j],land[i+1][j+1]]
        flag=True
        for k in surroundings:
            if point<=k:
                flag=False
                break
        if flag==True:
            peaks.append([i,j])
        else:
            flag2=True
            for k in surroundings:
                if point>=k:
                    flag2=False
                    break
            if flag2==True:
                valley.append([i,j])

print("INPUT LAND")
print(3*n*"-","---",sep="")
for i in land:
    print("| ",end="")
    for j in i:
        print(j," ",end="")
    print("|")
print(3*n*"-","---",sep="")

for i in peaks:
    land[i[0]][i[1]]="/\\"
for i in valley:
    land[i[0]][i[1]]="‿"
print("\nIDENTIFIED PEAKS")
print(3*n*"-","---",sep="")
for i in land:
    print("| ",end="")
    for j in i:
        if j!="/\\" and j!="‿":
            print("___",end="")
        else:
            print("_",j,end="",sep="")
    print("|")
print(3*len(land[0])*"-","---",sep="")
```
